```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_m = [0,0,-1]
y_m = [0,1,1]
x_p = [1,1,2]
y_p = [-1,-2,-2]
```

```python
plt.plot(x_m,y_m, linestyle="none", marker="o", color="blue", label="agua")
plt.plot(x_p,y_p, linestyle="none", marker="*", color="green", label="hierva")
plt.legend(loc=1)
plt.grid()
plt.show()
```

```python
x = [-1,2]
y = [-2,1]
```

```python
plt.plot(x_m,y_m, linestyle="none", marker="o", color="blue", label="agua")
plt.plot(x_p,y_p, linestyle="none", marker="*", color="green", label="hierva")
plt.plot(x, y, marker="p")
plt.legend(loc=7)
plt.grid()
plt.show()
```
