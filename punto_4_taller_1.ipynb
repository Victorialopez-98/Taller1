```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1=[2, 0, 2, 2]
y1=[0, -2, 2, -2]
x2=[0, -2, -2, -2]
y2=[2, 0, 2, -2]
```

```python
plt.plot(x1,y1, linestyle="none", marker="o", color="red")
plt.plot(x2,y2, linestyle="none", marker="*", color="black")
plt.grid()
plt.show()
```

```python
x = [-2, 2]
y = [-3, 3]
```

```python
plt.plot(x1,y1, linestyle="none", marker="o", color="red")
plt.plot(x2,y2, linestyle="none", marker="*", color="black")
plt.plot(x, y, marker=".")
plt.grid()
plt.show()
```

```python
m = (y[1] - y[0]) / (x[1] - x[0])
m
```
