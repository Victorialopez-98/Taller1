```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x1 = [-1,0,1]
y1 = [1,0,-1]
x2 = [1,0]
y2 = [0,1]
```

```python
plt.plot(x1,y1, linestyle="none", marker="o", color="red")
plt.plot(x2,y2, linestyle="none", marker="*", color="green")
plt.grid()
plt.show()
```

```python
x = [1, -0.5]
y = [-0.5, 1]
```

```python
plt.plot(x1,y1, linestyle="none", marker="o", color="red")
plt.plot(x2,y2, linestyle="none", marker="*", color="green")
plt.plot(x, y, marker=".")
plt.grid()
plt.show()
```
