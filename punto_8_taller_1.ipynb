```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_m = [0,-2,-2,-2,2]
y_m = [2,2,0,-2,2]
x_p = [0,2,2]
y_p = [-2,0,-2]
```

```python
plt.plot(x_m,y_m, linestyle="none", marker="o", color="red", label="tomates")
plt.plot(x_p,y_p, linestyle="none", marker="*", color="orange", label="naranjas")
plt.legend(loc=10)
plt.grid()
plt.show()
```

```python
x = [-2,3]
y = [-3,2]
```

```python
plt.plot(x_m,y_m, linestyle="none", marker="o", color="red", label="tomates")
plt.plot(x_p,y_p, linestyle="none", marker="*", color="orange", label="naranjas")
plt.plot(x, y, marker="p")
plt.legend(loc=4)
plt.grid()
plt.show()
```

```python
def funcion_y(x):
    return (1 * x) - 1
```

```python
lista_x = [0, -2, 2]
lista_y = [funcion_y(lista_x[0]), funcion_y(lista_x[1]), funcion_y(lista_x[2])]
```

```python
plt.plot(x_m,y_m, linestyle="none", marker="o", color="red", label="tomates")
plt.plot(x_p,y_p, linestyle="none", marker="*", color="orange", label="naranjas")
plt.plot(lista_x, lista_y, marker="p")
plt.legend(loc=4)
plt.grid()
plt.show()
```
